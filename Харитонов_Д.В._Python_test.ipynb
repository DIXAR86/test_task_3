    - для кратных пяти - печатать “Buzz”,
    - для кратных и трем, и пяти - печатать “FizzBuzz”

```python
# code here
for i in range (1, 101):
    if ((i/3)-(i//3) == 0) and ((i/5)-(i//5) == 0):
        print('FizzBuzz')
    elif (i/3)-(i//3) == 0:
        print('Fizz')
    elif (i/5)-(i//5) == 0:
        print('Buzz')
    else:
        print(i)
```

a) Аппроксимация. Применить метод наименьших квадратов (связка np.poly1d и np.polyfit) для заданных значений x и y на следующих моделях:

    - линейная

```python
import numpy as np
from matplotlib import pylab as plt
%matplotlib inline

x = np.arange(0, 1, 0.04)

# code here
a = -40
b = 30
c = 20
d = 7

y_lin = np.poly1d([c, b])(x)
y_quad = np.poly1d([a, b, c])(x)
y_cube = np.poly1d([a, b, c, d])(x)

y_lin_noise = np.random.normal(y_lin)
y_quad_noise = np.random.normal(y_quad)
y_cube_noise = np.random.normal(y_cube)
```

```python
# code_here
from matplotlib.pyplot import figure
import pylab 

figure(figsize=(8, 8))

plt.plot(x, y_lin, label = 'lin')
plt.plot(x, y_lin_noise, label = 'lin_noise')
plt.plot(x, y_quad, label = 'quad')
plt.plot(x, y_quad_noise, label = 'quad_noise')
plt.plot(x, y_cube, label = 'cube')
plt.plot(x, y_cube_noise, label = 'cube_noise')

pylab.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# функция расчета значения полинома. Порядок полинома передается в качестве параметра функции
def polynom (grade, x):
    counter = grade
    m = 1
    y_poly=0
    for i in range(counter):
        y_poly += (m*x**(m-1))
        m +=1
    return y_poly
```

```python
# a) Аппроксимация. Расчет значений 'y' для констатны и полиномов 5, 50 порядка и 'y_noise'.
# расчет нового значения y для построения апроксимированной линии
y_const = np.ones((25, ))*10
y_const_noise = np.random.normal(y_const)
y_const_approx = np.poly1d(np.polyfit(x, y_const_noise, 0))(x)

y_poly_5 = np.poly1d([a, b, c, d, 1])(x)
y_poly_5_noise = np.random.normal(y_poly_5)
y_poly_5_approx = np.poly1d(np.polyfit(x, y_poly_5_noise, 5))(x)

y_poly_50 = polynom(50,x)
y_poly_50_noise = np.random.normal(y_poly_50)
y_poly_50_approx = np.poly1d(np.polyfit(x, y_poly_50_noise, 50))(x)

y_lin_approx = np.poly1d(np.polyfit(x, y_lin_noise,1))(x)
y_quad_approx = np.poly1d(np.polyfit(x, y_quad_noise, 2))(x)
y_cube_approx = np.poly1d(np.polyfit(x, y_cube_noise, 3))(x)
```

```python
# строим графики

figure(figsize=(14, 4))
plt.plot(x, y_const, 'o', label='Исходная линия')
plt.plot(x, y_const_noise, '--', label='Линия с шумом')
plt.plot(x, y_const_approx, 'g', label='Аппроксимированная линия')
plt.xlabel('x')
plt.ylabel('y')

plt.title('1. Константа')
pylab.legend(loc='upper left')
plt.show()
# ===========


figure(figsize=(14, 4))
plt.plot(x, y_lin, 'o', label='Исходная линия')
plt.plot(x, y_lin_noise, '--', label='Линия с шумом')
plt.plot(x, y_lin_approx, 'g', label='Аппроксимированная линия')
plt.xlabel('x')
plt.ylabel('y')

plt.title('2. Линейная')
pylab.legend(loc='upper left')
plt.show()
# ===========


figure(figsize=(14, 4))
plt.plot(x, y_quad, 'o', label='Исходная линия')
plt.plot(x, y_quad_noise, '--', label='Линия с шумом')
plt.plot(x, y_quad_approx, 'g', label='Аппроксимированная линия')
plt.xlabel('x')
plt.ylabel('y')

plt.title('3. Квадратичная')
pylab.legend(loc='upper left')
plt.show()
# ===========


figure(figsize=(14, 4))
plt.plot(x, y_cube, 'o', label='Исходная линия')
plt.plot(x, y_cube_noise, '--', label='Линия с шумом')
plt.plot(x, y_cube_approx, 'g', label='Аппроксимированная линия')
plt.xlabel('x')
plt.ylabel('y')

plt.title('4. Кубическая')
pylab.legend(loc='upper left')
plt.show()
# ===========


figure(figsize=(14, 4))
plt.plot(x, y_poly_5, 'o', label='Исходная линия')
plt.plot(x, y_poly_5_noise, '--', label='Линия с шумом')
plt.plot(x, y_poly_5_approx, 'g', label='Аппроксимированная линия')
plt.xlabel('x')
plt.ylabel('y')

plt.title('5. 5го порядка')
pylab.legend(loc='upper left')
plt.show()
# ===========


figure(figsize=(14, 4))
plt.plot(x, y_poly_50, 'o', label='Исходная линия')
plt.plot(x, y_poly_50_noise, '--', label='Линия с шумом')
plt.plot(x, y_poly_50_approx, 'g', label='Аппроксимированная линия')
plt.xlabel('x')
plt.ylabel('y')

plt.title('6. Полином порядка 50')
pylab.legend(loc='upper left')
plt.show()

# ===========
```

    Вывести список периодов и континентов, о которых имеются данные.

a) Вывести количество записей о странах с продолжительностью жизни больше 58 или с ВВП больше 100000 за первые 4 периода

```python
# a) Вывести количество записей о странах с продолжительностью жизни больше 58 или 
# с ВВП больше 100 000 за первые 4 периода

"""
по данному пункту задачи возникли сомнения касательно ее интерпритации, а именно:
нужно было вывести просто строки по колонке 'gdpPercap' у которых значение больше 100 000
или же просуммировать значение по колонке 'gdpPercap' за выбранный период и уже после отобрать значение 
которые отвечают условиям.

Я интерпретировал задачу в пользу второго варианта (посчитал агрегированное значение за 4 первых периода
и потом уже применил фильтр)"""

# сделал копию исходно df
dfa = df.copy(deep=True)
```

```python
# установил значения для фильтров по году и средней продолжительности жизни
filtera1 = dfa['year'] < 1972
filtera2 = dfa['lifeExp'] > 58
```

```python
# применил оба фильтра.
# ответ на первую часть вопроса

dfa2 = dfa.where(filtera1 & filtera2).dropna(subset=['year'])

print('Выборка за первые 4 периода: \n\n'
     '    Кол-во записей о странах с продолжительность жизни более 58 лет: \n\n        -'
      ,len(dfa2),'записей;')
```

```python
# отфильтровал нужный временной период и не нужные строки удалил из копии df
dfa3 = dfa.where(filtera1).dropna(subset=['year'])

# удалил не нужные колонки
dfa3.drop(dfa3.columns[[1,2,3,4,6]], axis=1, inplace=True)

# сгруппировал значени по странам и просуммировал
dfa4 = dfa3.groupby(['country']).sum().sort_values('gdpPercap', ascending=False)
dfa4.head(10)
```

```python
# выполнил проверку на второе условие задачи
filtera3 = dfa4['gdpPercap'] > 100000

dfa5 = dfa4.where(filtera3).dropna(subset=['gdpPercap'])

print('Агрегированные значения за первые 4 периода: \n\n'
       '    Кол-во записей о странах с суммарным ВВП, за первые 4 периода, более 100 000 : \n\n        -',
      len(dfa5), 'значение;')
```

```python
# b) Вывести среднюю продолжительность жизни и количество стран 
# на континентах за каждый год. Колонки: continent, year, avg_lifeExp, countries

dfb = df.copy(deep=True)
```

```python
# сделал срез посчитав среднюю продолжительность жизни по годам
dfb2 = dfb.groupby(['continent','year']).mean().drop(columns=['gdpPercap','pop','marker'])

# сделал срез посчитав кол-во стран по годам
dfb3 = dfb.groupby(['continent','year']).sum().drop(columns=['gdpPercap','pop','lifeExp'])

# объединил два датафрейма
dfb4  = pd.merge(dfb2, dfb3, on=dfb2.index)

# в результате таблица с нужной информацией но полученный вид еще не соотвествует задаче
dfb4.head()
```

```python
# сделал временную копию колонки в которой содержиться информация т.к. нарямую почему-то колонка key_0 выдавала ошибку
dfb4['temp_c'] = dfb4['key_0']

# сделал два пустых списка в качестве заготовок для сбора информации
list_continent = []
list_year = []

# в цикле прошелся по временному столбцу и раскидал значения: страна, год по соответсвующим спискам
for i in range(len(dfb4)):
    list_continent.append(dfb4['temp_c'][i][0])
    list_year.append(dfb4['temp_c'][i][1])
```

```python
# вношу полученную информацию в датафрейм
dfb4['continent'] = list_continent
dfb4['year'] = list_year

# переименовываю колонки
dfb4 = dfb4.rename(columns={'lifeExp': 'avg_lifeExp', 'marker': 'qnty_countries'})

# расставляю колонки в соостветствии с условием задачи (заодно отбросил не нужные столбцы)
dfb5 = dfb4[['continent','year','avg_lifeExp','qnty_countries']]
```

```python
# c) Построить в одной плоскости графики динамики изменения средней продолжительности жизни на континентах по годам

dfc = dfb5.copy(deep=True)
```

```python
# положил в переменную для оси х года
x = pd.unique(dfc['year'])
x
```

```python
#  циклом создаю вложенные списки в которых по континентам группирую значения среднего уровня жизни по годам
continent_list = []
counter = -1
for i in range(len(dfc)):
    if i == 0:
        continent_list.append([dfc['continent'][i],[dfc['avg_lifeExp'][i]]])
        counter += 1
        
    elif dfc['continent'][i] == dfc['continent'][i-1]: 
        continent_list[counter][1].append(dfc['avg_lifeExp'][i])
    
    elif dfc['continent'][i] != dfc['continent'][i-1]:
        continent_list.append([dfc['continent'][i],[dfc['avg_lifeExp'][i]]])
        counter += 1
continent_list[0]
```
